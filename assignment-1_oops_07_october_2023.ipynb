```python
 '''Problem 1: Bank Account Create a class representing a bank account with attributes like account number, account holder name, 
 and balance. Implement methods to deposit and withdraw money from the account.'''
    
import logging as log
log.basicConfig(filename="bankAccount.log", level=log.DEBUG, format="%(asctime)s - %(levelname)s - %(message)s")
class bankAccount:

    try:
        log.info("Creating constructor for bank account with account number, account holder name and balance as attributes.")
        def __init__(self, account_number, account_holder_name, balance):
            self.account_number = account_number
            self.account_holder_name = account_holder_name
            self.balance = balance
    except Exception as e:
        log.error(e)

    try:
        def deposit(self, amount_to_be_deposited):
            log.info("Inside 'deposit()' method for depositing money.")
            self.balance = (self.balance + amount_to_be_deposited)
            log.info(f"Balance after depositing : {self.balance}")
    except Exception as e:
        log.error(e)

    try:
        def withdraw(self, amount_to_be_withdrawn):
            log.info("Inside 'withdraw()' method for withdrawing money.")
            self.balance = (self.balance - amount_to_be_withdrawn)
            log.info(f"Balance after withdrawal : {self.balance}")
    except Exception as e:
        log.error(e)

try:
    log.info("Creating object(user1) for class bankAccount.")
    user1 = bankAccount(123456789, "Amit Raj", 3300)
    log.info("Object(user1) for class bankAccount is created.")
except Exception as e:
    log.error(e)

try:
    log.info(f"user1 account number - {user1.account_number}")
except Exception as e:
    log.error(e)

try:
    log.info(f"user1 account holder name - {user1.account_holder_name}")
except Exception as e:
    log.error(e)

try:
    log.info(f"user1 balance - {user1.balance}")
except Exception as e:
    print(e)

try:
    log.info("Calling 'deposit()' method")
    user1.deposit(500)
except Exception as e:
    log.error(e)

try:
    log.info("Calling 'withdraw()' method")
    user1.withdraw(700)
except Exception as e:
    log.error(e)
```

```python
'''Problem 2: Employee Management Create a class representing an employee with attributes like employee ID, name, and salary. 
Implement methods to calculate the yearly bonus and display employee details.'''

import logging as log
log.basicConfig(filename="employee_management.log", level=log.DEBUG, format="%(asctime)s - %(levelname)s - %(message)s")
class employee_management:
    try:
        log.info("Creating constructor for class with attributes as employee id, employee name and employee salary.")
        def __init__(self, emp_id, emp_name, emp_salary):
            self.employee_id = emp_id
            self.employee_name = emp_name
            self.employee_salary = emp_salary
    except Exception as e:
        log.error(e)

    try:
        def employee_details(self):
            log.info("Inside 'employee_details()' method")
            log.info(f"Employee id is {self.employee_id}, employee name is {self.employee_name} and employee salary is {self.employee_salary}.")
    except Exception as e:
        log.error(e)

    try:
        def bonus(self):
            log.info("Inside 'bonus()' method")
            if self.employee_salary >= 15000:
                log.info("If employee salary is greater than or equal to 15000")
                bonus = (20 / 100 * self.employee_salary)
                log.info(f"Bonus is {bonus}.")
                log.info("Total salary in hand :")
                total_salary_in_hand = (self.employee_salary + bonus)
                log.info(f"Thus, total hand in salary is {total_salary_in_hand}.")
            else:
                log.info("bonus when salary is less than 15000")
                bonus = (10 / 100 * self.employee_salary)
                log.info(f"Bonus is {bonus}.")
                log.info("Total salary in hand :")
                total_salary_in_hand = (self.employee_salary + bonus)
                log.info(f"Thus, total hand in salary is {total_salary_in_hand}.")
    except Exception as e:
        log.error(e)

try:
    log.info("Creating an object(employee_1) for the employee_management class.")
    employee_1 = employee_management(1, "Amit Raj", 17000)
    log.info("Object(employee_1) for class employee_management is created.")
except Exception as e:
    log.error(e)

try:
    log.info(f"employee_1 id is - {employee_1.employee_id}")
except Exception as e:
    log.error(e)

try:
    log.info(f"employee_1 name is - {employee_1.employee_name}")
except Exception as e:
    log.error(e)

try:
    log.info(f"employee_1 salary is - {employee_1.employee_salary}")
except Exception as e:
    log.error(e)

try:
    log.info("Calling 'employee_details()' method :")
    employee_1.employee_details()
except Exception as e:
    log.error(e)

try:
    log.info("Calling 'bonus()' method :")
    employee_1.bonus()
except Exception as e:
    log.error(e)
```

```python
'''Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height.
Implement methods to calculate the area and perimeter of the shape.'''

import logging as log
log.basicConfig(filename="shapeCalculation.log", level=log.DEBUG, format="%(asctime)s - %(levelname)s - %(message)s")
class shapeCalculation:

    try:
        log.info("Constructor for the class 'shapeCalculation' with attributes like 'length' and 'width'.")
        def __init__(self, length, width):
            self.length = length
            self.width = width
    except Exception as e:
        log.error(e)

    try:
        def area_of_rectangle(self):
            log.info("Inside 'area_of_rectangle()' method")
            area = (self.width * self.length)
            log.info(f"The area of the rectangle is {area}cm².")
    except Exception as e:
        log.error(e)

    try:
        def perimeter_of_rectangle(self):
            log.info("Inside 'perimeter_of_rectangle()' method")
            perimeter = 2 * (self.length + self.width)
            log.info(f"The perimeter of the rectangle is {perimeter}cm.")
    except Exception as e:
        log.error(e)

try:
    log.info("Creating object(rectangle_1) for the class 'shapeCalculation'")
    rectangle_1 = shapeCalculation(10, 15)
    log.info("Created object(rectangle_1) for the class 'shapeCalculation'")
except Exception as e:
    log.error(e)

try:
    log.info(f"Length of the rectangle - {rectangle_1.length}cm")
except Exception as e:
    log.error(e)

try:
    log.info(f"Width of the rectangle - {rectangle_1.width}cm")
except Exception as e:
    log.error(e)

try:
    log.info("Calling 'area_of_rectangle()' method")
    rectangle_1.area_of_rectangle()
except Exception as e:
    log.error(e)

try:
    log.info("Calling 'perimeter_of_rectangle()' method")
    rectangle_1.perimeter_of_rectangle()
except Exception as e:
    log.error(e)
```

```python
'''Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates.
Implement methods to add tasks, mark tasks as completed, and display pending tasks.'''

import logging as log
log.basicConfig(filename="todoList.log", level=log.DEBUG, format="%(asctime)s - %(levelname)s - %(message)s")
class todoList:

    try:
        log.info("Constructor for the class 'todoList' with attributes as 'tasks' and 'due_date'")
        def __init__(self, tasks, due_date):
            self.tasks = tasks
            self.due_date = due_date
    except Exception as e:
        log.error(e)
    
    try:
        def todo_list(self):
            log.info("Inside 'todo_list()' method")
            todo_list = self.tasks
            end_of_todo = 'n'
            while end_of_todo != 'y':
                todo_item = input("Enter task to do :\n")
                todo_list.append(todo_item)
                end_of_todo = input("Enter 'n' if you want to continue adding tasks or enter 'y' to stop :\n")
            log.info(f"My todo list is {todo_list}")
    except Exception as e:
        log.error(e)

    try:
        def completed_task(self):
            log.info("Inside 'completed_task()' method")
            import random
            tasks_completed = self.tasks.pop(random.randint(0, (len(self.tasks)-1)))
            log.info(f"Task completed = {tasks_completed}")
    except Exception as e:
        log.error(e)

    try:
        def pending_tasks(self):
            log.info("Inside 'pending_tasks()' method")
            pending_tasks = self.tasks
            log.info(f"Pending tasks are : {pending_tasks}")
    except Exception as e:
        log.error(e)

try:
    log.info("Creating an object(task1) for the class 'todoList'")
    task1 = todoList(["wash clothes", "clean floor"], "11/11/2023")
    log.info("Created an object(task1) for the class 'todoList'")
except Exception as e:
    log.error(e)

try:
    log.info(f"Tasks for 'task1' - {task1.tasks}")
except Exception as e:
    log.error(e)

try:
    log.info(f"Due date to complete all the tasks is : {task1.due_date}")
except Exception as e:
    log.error(e)

try:
    log.info("Calling 'todo_list()' method")
    task1.todo_list()
except Exception as e:
    log.error(e)

try: 
    log.info("Calling 'completed_task()' method")
    task1.completed_task()
except Exception as e:
    log.error(e)

try:
    log.info("Calling 'pending_tasks()' method")
    task1.pending_tasks()
except Exception as e:
    log.error(e)
```
